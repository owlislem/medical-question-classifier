```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Dataset
data = pd.DataFrame({
    "question":[
        "Which artery supplies the heart muscle?",
        "What is the normal blood pressure range?",
        "What causes myocardial infarction?",
        "How is coronary artery disease diagnosed?",
        "Name the function of the liver",
        "Which organ detoxifies blood?",
        "What enzyme is produced by the pancreas?",
        "What is hepatitis?",
        "What is MRI used for?",
        "Which imaging technique uses magnetic fields?",
        "Which scan uses X-rays?",
        "What does CT stand for in medical imaging?",
        "Which nerve controls eye movement?",
        "Which nerve controls facial muscles?",
        "What part of the brain controls balance?",
        "What is the role of the spinal cord?"
    ],
    "label":[
        "cardiology",
        "cardiology",
        "cardiology",
        "cardiology",
        "gastroenterology",
        "gastroenterology",
        "gastroenterology",
        "gastroenterology",
        "radiology",
        "radiology",
        "radiology",
        "radiology",
        "neurology",
        "neurology",
        "neurology",
        "neurology"
    ]
})

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    data["question"],
    data["label"],
    test_size=0.25,
    random_state=42
)

print("Train size:", len(X_train))
print("Test size:", len(X_test))

# Text vectorization
vectorizer = TfidfVectorizer(ngram_range=(1,2))
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

# Predictions
preds = model.predict(X_test_vec)

# Evaluation
print(classification_report(y_test, preds, zero_division=0))
```
